Recreate Interest vs Principal Graph (Practice)

```python
import pandas as pd


url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vSdgsW4J8l9V8UlKLuwdYdqyemwkAuTHzS4s6plb-BQIJ8DhI8BkNDvO_Ei3M4lBJElVEaueyOg1Jkj/pub?output=csv"
df = pd.read_csv(url)
df.head()
```

```python
import matplotlib.pyplot as plt


xAxis = df['Month']
yAxis = df['3% Cumulative Interest']
zAxis = df['5% Cumulative Interest']

# Adding labels for each line and a legend

fig , ax = plt.subplots()

ax.plot(xAxis, yAxis, label = "30 years 3% APR")
ax.plot(xAxis, zAxis, label = "30 years 5% APR")
ax.legend()
# Using ax.set to customize
ax.set(ylabel= "Dollars ($)", xlabel = "Months", title = "Cumulative Interest on 400K loan ");


```
