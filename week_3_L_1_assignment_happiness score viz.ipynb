```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',100)
```

```python
# This filepath works if you have added the Coding Dojo file structure to your Drive
fpath = "/content/drive/MyDrive/CodingDojo/01-Fundamentals/happiness_score.csv"
df = pd.read_csv(fpath)
df.head()

```

## 1. Which region have highest happiness score, and which have lowest


```python
# This filepath works if you have added the Coding Dojo file structure to your Drive



region_mean_happiness = df.groupby('Region')['Happiness Score'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=region_mean_happiness.index, y=region_mean_happiness.values, palette="viridis")
plt.xticks(rotation=45, ha='right')
plt.title('Mean Happiness Score by Region')
plt.xlabel('Region')
plt.ylabel('Mean Happiness Score')
plt.show()


```

## 2. Which regions have countries with extream (outliers) happiness score?


```python
sns.boxplot(data=df , x='Region',y='Happiness Score' )
plt.xticks(rotation=45, ha='right')
plt.title('Mean Happiness Score by Region')
plt.xlabel('Region')
plt.ylabel('Mean Happiness Score')
plt.show()
```

## 3. Show how the global happiness score change over time


```python
import matplotlib.ticker as ticker

plt.figure(figsize=(10, 6))

ax = sns.lineplot(x='Year', y='Happiness Score', data=df, marker='o', estimator='mean', ci=None)

plt.title('Global Happiness Score Over Time')
plt.xlabel('Year')
plt.ylabel('Average Happiness Score')
plt.grid(True)

ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.show()
```

## 4. How does the Happiness Score change by Region?

```python
# Assuming 'Year' is available in your dataset
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Happiness Score', data=df, marker='o',hue='Region' ,estimator='mean', ci=None)
plt.title('Global Happiness Score Over Time')
plt.xlabel('Year')
plt.ylabel('Average Happiness Score')
plt.grid(True)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```
