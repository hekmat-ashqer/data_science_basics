```python
import pandas as pd

# path for the insurance csv file
file_path = '/content/drive/MyDrive/CodingDojo/01-Fundamentals/Week02/Lectures/NFLX.csv'

# load csv file as data frame

netflix_df = pd.read_csv(file_path)
netflix_df

```

```python
netflix_df.describe()
```

```python
# Filter columns based on numerical data types (int or float)
numerical_columns = netflix_df.select_dtypes(include=['int', 'float'])

# Count the number of numerical columns
numerical_column_count = len(numerical_columns.columns)
numerical_column_count
```

```python
# Filter columns based on object or category data types (categorical data)
categorical_columns = netflix_df.select_dtypes(include=['object', 'category'])

# Count the number of categorical columns
categorical_column_count = len(categorical_columns.columns)
categorical_column_count
```

```python
column_median = netflix_df.median()
column_median
```

```python
#column_means = round(netflix_df.mean(),2)
column_means = round(netflix_df.mean(),2)
column_means
```

```python
column_count = netflix_df.value_counts()
column_count
```

```python
# Convert the 'Date' column to datetime
netflix_df['Date'] = pd.to_datetime(netflix_df['Date'])

# Group by month and calculate the sum of 'Value' for each month
monthly_data = netflix_df.groupby(netflix_df['Date'].dt.to_period('M')).mean()



monthly_data
```
